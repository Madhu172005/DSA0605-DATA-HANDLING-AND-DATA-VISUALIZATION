```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
df = pd.read_csv("dataset.csv")

# Calculate total sales for each product
sales_by_product = df.groupby('Region')['Sales'].sum().reset_index()

# Plotting pie chart
plt.figure(figsize=(10, 6))
plt.pie(sales_by_product['Sales'], labels=sales_by_product['Region'], autopct='%1.1f%%', startangle=90, colors=plt.cm.tab20.colors)
plt.title('Sales Distribution by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
df = pd.read_csv("dataset.csv")

# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Region'], bins=len(df['Region'].unique()), color='brown', edgecolor='black')
plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("dataset.csv")

# Group by customer ID and count the number of unique products purchased by each customer
customer_product_count = df.groupby('Customer_ID')['Product_ID'].nunique()

# Find the top 5 customers with the highest product count
top_5_customers = customer_product_count.sort_values(ascending=False).head(5)

# Display the top 5 customers
print("Top 5 Customers Who Purchased the Most Products:")
print(top_5_customers)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("dataset.csv")

# Group by product ID and sum the sales for each product
product_sales = df.groupby('Product_ID')['Sales'].sum()

# Find the top 7 best-selling products
top_7_products = product_sales.sort_values(ascending=False).head(7)

# Display the top 7 best-selling products
print("Top 7 Best-Selling Products:")
print(top_7_products)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("dataset.csv")

category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Plotting the graph
plt.figure(figsize=(12, 6))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("dataset.csv")

# Group by region and sum the sales for each region
region_sales = df.groupby('Region')['Sales'].sum()

# Find the region with the highest sales
max_sales_region = region_sales.idxmax()
max_sales_value = region_sales.max()

# Display the region with the highest sales
print(f"Region with the highest sales: {max_sales_region} (Total Sales: {max_sales_value})")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("dataset.csv")

# Group by region and sum the sales for each region
region_sales = df.groupby('Region')['Sales'].sum()

# Find the region with the lowest sales
min_sales_region = region_sales.idxmin()
min_sales_value = region_sales.min()

# Display the region with the lowest sales
print(f"Region with the lowest sales: {min_sales_region} (Total Sales: {min_sales_value})")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("dataset.csv")

# Group by category and region, and sum the sales
grouped_df = df.groupby(['Category', 'Region'])['Sales'].sum().unstack()

# Create the stacked bar graph
plt.figure(figsize=(12, 6))
grouped_df.plot(kind='bar', stacked=True)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Stacked Bar Graph of Sales by Category and Region')
plt.legend(title='Region', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("dataset.csv")

# Create the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Segment', y='Region', data=df)
plt.xlabel('Segment')
plt.ylabel('Region')
plt.title('Box Plot of Segment and Region')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("dataset.csv")

# Group by category and sum the sales
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

# Sort the data by sales in descending order
category_sales = category_sales.sort_values('Sales', ascending=False)

# Create the funnel graph
fig = px.funnel(category_sales, x='Sales', y='Category')

# Update layout for better readability
fig.update_layout(title='Funnel Graph of Sales by Category',
                  xaxis_title='Sales',
                  yaxis_title='Category',
                  plot_bgcolor='rgba(0, 0, 0, 0)')

# Show the plot
fig.show()
```
